Lista de precios

***Corte de pelo***
> Gato: 20

> Perro: 30

> Conejo: 15

***Baño***


> Gato: 15


> Perro: 20


> Conejo: 10

***Desparasitación***


> Gato: 35

> Perro: 50

> Conejo: 25

```python
# %%pycodestyle
import argparse
import sys


def main():
    parser = argparse.ArgumentParser()
    parser.add_argument('--tratamiento', type=str, default='null', help='Ingrese el tipo de tratamiento a utilizar')
    parser.add_argument('--animal', type=str, default='null', help='Ingrese el tipo de animal a realizar el tratamiento')
    args = parser.parse_args()
    sys.stdout.write(str(EvalucionTratamiento(args)))


def EvalucionTratamiento(args):
    if (args.tratamiento == "") & (args.animal == ""):
        print('Tiene que especificar el tratamiento y el animal')
    elif args.tratamiento == "Corte de pelo":
        if args.animal == "Gato":
            print(f'El costo por el tratamiento de {args.tratamiento} de su {args.animal} es: S/. 20')
        elif args.animal == "Perro":
            print(f'El costo por el tratamiento de {args.tratamiento} de su {args.animal} es: S/. 30')
        elif args.animal == "Conejo":
            print(f'El costo por el tratamiento de {args.tratamiento} de su {args.animal} es: S/. 15')
        else:
            print('Tiene que especificar el animal')
    elif args.tratamiento == "Baño":
        if args.animal == "Gato":
            print(f'El costo por el tratamiento de {args.tratamiento} de su {args.animal} es: S/. 15')
        elif args.animal == "Perro":
            print(f'El costo por el tratamiento de {args.tratamiento} de su {args.animal} es: S/. 20')
        elif args.animal == "Conejo":
            print(f'El costo por el tratamiento de {args.tratamiento} de su {args.animal} es: S/. 10')
        else:
            print('Tiene que especificar el animal')
    elif args.tratamiento == "Desparacitación":
        if args.animal == "Gato":
            print(f'El costo por el tratamiento de {args.tratamiento} de su {args.animal} es: S/. 35 ')
        elif args.animal == "Perro":
            print(f'El costo por el tratamiento de {args.tratamiento} de su {args.animal} es: S/. 50 ')
        elif args.animal == "Conejo":
            print(f'El costo por el tratamiento de {args.tratamiento} de su {args.animal} es: S/. 25 ')
        else:
            print('Tiene que especificar el animal')
    else:
        print('Tiene que especificar el tratamiento')


if __name__ == '__main__':
    main()
```

```python
# %%pycodestyle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

covid = pd.read_csv("/content/drive/MyDrive/archivos_python/positivos_covid.csv", delimiter=';')
covid["MES_RESULTADO"] = covid.FECHA_RESULTADO.apply(lambda x: str(x)[:6])
covid["DEP_LIMA"] = covid.DEPARTAMENTO.apply(lambda x: 1 if x == 'LIMA' else 0)

df = covid.groupby(["MES_RESULTADO", "DEP_LIMA"])[["id_persona"]].aggregate('count').reset_index()
# grafica de linea
plt.rcParams["figure.figsize"] = (20, 6)
sns.lineplot(x="MES_RESULTADO", y="id_persona", hue="DEP_LIMA", data=df)
plt.xticks(rotation=90)
plt.show()
```

**Observacion** : La linea *naranja* son los casos de covid positivo de ***LIMA*** y la linea azul son los demas departamentos.


4. (3 puntos) En el dataset de Titanic train enlace visto en clase, realice un gráfico de cajas sobre la variable Fare (Costo del ticket), ¿qué observa?, ¿Hay outliers?, ¿Qué tratamiento recomendaría?

```python
# %%pycodestyle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# pd.__version__
# !pip3 install pandas==1.3.5 -vvv -forced
```

```python
# %%pycodestyle
# leyendo archivos CSV
pddf = pd.read_csv('/content/drive/MyDrive/archivos_python/train_titanic.csv')
# Pandas dataframe
sns.set_theme(context='talk', style="whitegrid", font_scale=1)
ax = sns.boxplot(data=pddf.Fare, orient="h")
plt.xticks(list(range(-25, 550, 25)), rotation=90)
plt.show()
```

Observacion: que los datos atipicos(outliers) afectan la buena interpretacion del diagrama de cajas, tamnien que hay una ligera asimetria positiva.
Tratamiento:para una mejor interpretacion del diagrama de cajas eliminar los datos outliers.

```python
# %%pycodestyle
Me = np.percentile(pddf.Fare.dropna(), 50)
Q1 = np.percentile(pddf.Fare.dropna(), 25)
Q3 = np.percentile(pddf.Fare.dropna(), 75)
RIC = Q3-Q1
LI = Q1-1.5*RIC
LS = Q3+1.5*RIC
pddf['Fare_1'] = pddf.Fare.apply(lambda x: LS if x > LS else x)
sns.set_theme(context='talk', style="whitegrid", font_scale=0.7)
ax = sns.boxplot(data=pddf.Fare_1, orient="h")
plt.xticks(list(range(-5, 70, 5)), rotation=90)
plt.show()
```

Observamos: Despues de quitarle los datos outliers, se observa que los datos no son dispersos, y la distribucion de los mismos datos son asimetricos positiva, Los costos de ticket en el titanic van desde los 0 a 55 dólares, El 50% de los tickets tienen precios entre los 8 y 31 dólares.
